Preparing Trailer_v3_dataset (The complete dataset)

```python
data_df = pd. read_csv('/home/app/src/CSV/dataset_df.csv', dtype={"shot": str})

data_df.drop(columns=['scale_val'], inplace=True)
data_df.drop(columns = ['move_val'], inplace=True)

train_path = '/home/app/src/data/shot-type-dataset/trailer_v3/train/'
data_df['video_name'] = train_path + 'shot_' + data_df['movie'] + '_' + data_df['shot'] + '.mp4'
train_df = data_df[(data_df['dataset']=='train') | (data_df['dataset']=='val') ]
train_df = train_df[['video_name', 'move_label']]
train_df = train_df.rename(columns ={'move_label': 'tag'})


test_path = '/home/app/src/data/shot-type-dataset/trailer_v3/test/'
data_df['video_name'] = test_path + 'shot_' + data_df['movie'] + '_' + data_df['shot'] + '.mp4'
test_df = data_df[data_df['dataset']=='test']
test_df = test_df[['video_name', 'move_label']]
test_df = test_df.rename(columns ={'move_label': 'tag'})
```

Deleting Multi-movement Tag


```python
train_df = train_df[train_df.tag != 'Multi_movement']
train_df = train_df.reset_index(drop = True) 
test_df = test_df[test_df.tag != 'Multi_movement']
test_df = test_df.reset_index(drop = True) 
```

Reducing Dataset

```python
train_df_bal_static = train_df[train_df.tag == 'Static'][0:320]
train_df_bal_motion = train_df[train_df.tag == 'Motion'][0:320]

len_push_tag = train_df.tag[train_df.tag == 'Push'].shape[0]
len_pull_tag = train_df.tag[train_df.tag == 'Pull'].shape[0]

train_df_bal_push = train_df[train_df.tag == 'Push'][0:len_push_tag-1243]
train_df_bal_pull = train_df[train_df.tag == 'Pull'][0:len_pull_tag]

train_df= pd.concat([train_df_bal_static, train_df_bal_motion, train_df_bal_pull, train_df_bal_push], axis=0)
```

Mixing the datasets

```python
train_df=train_df.sample(frac=1).reset_index(drop=True)
```

```python
train_df
```

```python
test_df_bal_static = test_df[test_df.tag == 'Static'][0:80]
test_df_bal_motion = test_df[test_df.tag == 'Motion'][0:80]

len_push_tag = test_df.tag[test_df.tag == 'Push'].shape[0]
len_pull_tag = test_df.tag[test_df.tag == 'Pull'].shape[0]

test_df_bal_push = test_df[test_df.tag == 'Push'][0:len_push_tag-140]
test_df_bal_pull = test_df[test_df.tag == 'Pull'][0:len_pull_tag]

test_df= pd.concat([test_df_bal_static, test_df_bal_motion, test_df_bal_pull, test_df_bal_push], axis=0)
```

Mixing the datasets

```python
test_df=test_df.sample(frac=1).reset_index(drop=True)
```
